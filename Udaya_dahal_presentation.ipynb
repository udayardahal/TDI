```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df1 = pd.read_csv('natl2018us.csv',nrows=1000000,low_memory=False)
```

```python
df1.keys()
```

```python
r = requests.get('https://data.nber.org/natality/2018/natl2018.sas')
```

```python
text = r.text.split('\n')
```

```python
labels = {}
for i in text:
    if i.strip() != '':
       if i.split()[0] == 'attrib':
           temp = (i.split(';')[0])
           k = temp.split()[1]
           v = temp.split('"')[-2]
           labels[k]=v
```

```python
features = ['Cigarettes Before Pregnancy 00-97 Number of cigarettes daily',
 'Cigarettes 1st Trimester 00-97 Number of cigarettes daily',
 'Cigarettes 2nd Trimester 00-97 Number of cigarettes daily',
 'Cigarettes 3rd Trimester 00-97 Number of cigarettes daily',
 'Cigarettes Before Pregnancy Recode 0 Nonsmoker',
 'Cigarettes 1st Trimester Recode 0 Nonsmoker',
 'Cigarettes 2nd Trimester Recode 0 Nonsmoker',
 'Cigarettes 3rd Trimester Recode 0 Nonsmoker',
 'Cigarette Recode (Revised) Y Yes',
 "Mother's Height in Total Inches 30-78 Height in inches",
 'Body Mass Index 13.0-69.9 Body Mass Index',
 'Body Mass Index Recode 1 Underweight <18.5',
 'Pre-pregnancy Weight Recode 075-375 Weight in pounds',
 'Delivery Weight Recode 100-400 Weight in pounds',
 'Weight Gain 00-97 Weight gain in pounds',
 'Weight Gain Recode 1 Less than 11 pounds',
 'Pre-pregnancy Diabetes Y Yes',
 'Gestational Diabetes Y Yes',
 'Pre-pregnancy Hypertension Y Yes',
 'Gestational Hypertension Y Yes',
 'Hypertension Eclampsia Y Yes',
 'Previous Preterm Birth Y Yes',
 'Infertility Treatment Used Y Yes',
 'Fertility Enhancing Drugs Y Yes',
 'Asst. Reproductive Technology Y Yes',
 'Previous Cesarean Y Yes',
 'No Risk Factors Reported 1 True',
 'Gonorrhea Y Yes',
 'Syphilis Y Yes',
 'Chlamydia Y Yes',
 'Hepatitis B Y Yes',
 'Hepatitis C Y Yes',
 'Steroids Y Yes',
 'Antibiotics Y Yes',
 'Chorioamnionitis Y Yes',
 'No Characteristics of Labor Reported 1 True']
```

```python
feature_ids = []
for k,v in labels.items():
    if v in features:
        feature_ids.append(k)
```

```python
df2 = df1[feature_ids]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import re
```

```python
new_key = [s  for s in list(df2.keys()) if (re.findall('_r',s) != ['_r'])]
```

```python
df3 = df2[new_key]
df3_norm = df3[new_key[:-1]]
df3_norm = (df3_norm - df3_norm.min())/(df3_norm.max() - df3_norm.min())
```

```python
df3.head()
```

```python
fig = plt.subplots(figsize=(16,8))
sns.heatmap(df3_norm.corr(),annot=True,fmt='.2f')
```

```python
X = df3_norm
```

```python
y = np.zeros(len(df3_norm))
```

```python
pca = PCA(n_components = 25)
```

```python
X_train,X_test ,y_train, y_test = train_test_split(X,y,
                                                   test_size=100,
                                                   random_state=0,
                                                  shuffle = True,
                                                  stratify = y)
```
